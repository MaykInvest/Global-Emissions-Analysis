##### Check for missing values on the DataFrame:

```python
df.isna().sum()  # You can also use df.isnull().sum() — both do the same thing
```

```python
# Drop the all‑null ISO2 column. SO2 has 100% missing values → i have to drop this column, it won’t be useful.
df = df.drop(columns=['ISO2', 'ObjectId'])
```

```python
df.columns
```

```python
df.dtypes
```

```python
# Identify all columns that are NOT quarter columns (i.e., identifiers)
id_vars = [col for col in df.columns if not col.startswith(('2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023','2024','2025'))]

# Quarter columns are the rest
quarter_cols = [col for col in df.columns if col not in id_vars]

# Melt the dataframe
df_long = pd.melt(df,
                  id_vars=id_vars,
                  value_vars=quarter_cols,
                  var_name='Quarter',
                  value_name='Emissions')

# Check the result
print("New shape:", df_long.shape)
print(df_long.head())
```

```python
# Convert Quarter string to datetime (using first day of the quarter)
df_long['Date'] = pd.to_datetime(df_long['Quarter'])

# Check the result
print(df_long[['Quarter', 'Date']].head())
```

```python
# Check for missing emissions
print("Missing emissions:", df_long['Emissions'].isnull().sum())

# Check unique values in key columns
print("Unique Indicators:", df_long['Indicator'].nunique())
print(df_long['Indicator'].value_counts().head())

print("\nUnique Industries:", df_long['Industry'].nunique())
print(df_long['Industry'].value_counts().head())

print("\nUnique Gas Types:", df_long['Gas Type'].nunique())
print(df_long['Gas Type'].value_counts().head())

print("\nUnique Seasonal Adjustment:", df_long['Seasonal Adjustment'].unique())
```

No missing emissions – that’s excellent and simplifies the analysis.

One indicator – we are working with the quarterly greenhouse gas emissions accounts. Good, no need to filter multiple indicators.

10 industries – the counts per industry are balanced (most appear 15,250 times except Transportation and Storage and Total Households, which are slightly different). This is expected because some industries may not exist in all countries or all gas types.

5 gas types – notice that Fluorinated gases has fewer rows (23,668) than the others (30,500). That likely means not all country–industry–seasonality combinations report fluorinated gases. That’s realistic.

Two seasonal adjustments – you have both “Not Seasonally Adjusted” and “Seasonally Adjusted” versions for each combination. You’ll need to decide which one to use for trend analysis. Often, seasonally adjusted series are preferred to see underlying trends without quarterly fluctuations.

```python
# after clear the data, it's always important to save a copy to avoid lose the data
dfc = df_long.copy()
```
